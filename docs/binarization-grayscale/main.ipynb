# Criando Funções para Manipular a Imagem

```python
def rgb_to_gray_scale(image: np.ndarray) -> np.ndarray:
    """
    Convert an RGB image to gray scale.
    
    - Args:
        - image: np.ndarray: An RGB image.
    
    - Returns:
        - np.ndarray: The gray scale image.
    """
    
    height, width, _ = image.shape
    
    for x in range(width):
        
        for y in range(height):
            
            r, g, b = image[y, x]
            
            gray = (r + g + b) / 3
            
            image[y, x] = gray
            
    return image
```

```python
def gray_scale_to_binary(image: np.ndarray, threshold: float = 0.6) -> np.ndarray:
    """
    Convert a gray scale image to binary.
    
    - Args:
        - image: np.ndarray: A gray scale image.
        - threshold: int: The threshold value.
        
    - Returns:
        - np.ndarray: The binary image.
    """
    # Ensure the image is 2D
    if len(image.shape) == 3:
        image = image[:, :, 0]
        
    if image.max() > 1:
        image = image / 255
    
    height, width = image.shape

    
    for x in range(width):
        for y in range(height):

            if image[y, x] > threshold:
                image[y, x] = 255
            else:
                image[y, x] = 0
                
    return image
```

# Aplicando as Funções Nas Imagens

```python
gray_image = rgb_to_gray_scale(image.copy())
```

```python
binary_image = gray_scale_to_binary(gray_image.copy())
```
